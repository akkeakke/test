```python
import re
import pandas as pd
import ast
import json
import dask.dataframe as dd
from requests_oauthlib import OAuth1Session
import sys, time, calendar
```

```python
ck = "#############èªè¨¼csafj;dlja####################"
cs = "###############èªè¨¼ã‚³ãƒ¼ãƒ‰##################"
at = "#################èªè¨¼ã‚³ãƒ¼ãƒ‰################"
ats = "###############èªè¨¼ã‚³ãƒ¼ãƒ‰#################"
twitter = OAuth1Session(ck, cs, at, ats)  #OAuthèªè¨¼
```

```python
geocode = '' #ç·¯åº¦çµŒåº¦ãªã—
```

ãƒ»ã€æ¤œç´¢æ¡ä»¶ã€‘ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰ã€å–å¾—ãƒ„ã‚¤ãƒ¼ãƒˆæ•°ã€è¨€èªã€çµæœã®ã‚¿ã‚¤ãƒ—ã€æ—¥æ™‚æ¡ä»¶ã€ãƒ„ã‚¤ãƒ¼ãƒˆIDæ¡ä»¶

```python
q =  '' #æ¤œç´¢ã—ãŸã„ã‚­ãƒ¼ãƒ¯ãƒ¼ãƒ‰
count = 100#æ¡ä»¶ï¼š1ï½100 180req/15min
lang = 'ja'#è¨€èª
result_type = 'recent'#recent:æœ€æ–°ã€popular:äººæ°—ã®ãƒ„ã‚¤ãƒ¼ãƒˆ
until = ''  #æŒ‡å®šã—ãŸæ—¥ä»˜ã¾ã§å–ã‚‹
max_id = ''#å–å¾—ã—ãŸã„æœ€æ–°ã®ãƒ„ã‚¤ãƒ¼ãƒˆID(ã“ã“ã‹ã‚‰æœ€æ–°100ä»¶)

url = "https://api.twitter.com/1.1/search/tweets.json" #ãƒªã‚¯ã‚¨ã‚¹ãƒˆURL
```

ãƒ»å„ãƒªã‚¯ã‚¨ã‚¹ãƒˆã”ã¨ã«ãƒ„ã‚¤ãƒ¼ãƒˆã‚’å–å¾—ã™ã‚‹é–¢æ•°ã‚’ä½œæˆ

```python
def get_tweet(url,params):
    req = twitter.get(url, params = params)
    name = [] #ãƒ¦ãƒ¼ã‚¶ãƒ¼ã®ãƒ‹ãƒƒã‚¯ãƒãƒ¼ãƒ  
    s_name = [] #ãƒ¦ãƒ¼ã‚¶ãƒ¼ç™»éŒ²å
    tid = [] #ãƒ„ã‚¤ãƒ¼ãƒˆIDï¼ˆãƒ¦ãƒ‹ãƒ¼ã‚¯ï¼‰
    text = [] #ãƒ„ã‚¤ãƒ¼ãƒˆå†…å®¹
    urls = [] #ãƒ„ã‚¤ãƒ¼ãƒˆurl
    geo = [] #ãƒ„ã‚¤ãƒ¼ãƒˆã®ä½ç½®æƒ…å ±
    c_at = [] #ãƒ„ã‚¤ãƒ¼ãƒˆæ—¥æ™‚
    ret_count = [] #ãƒªãƒ„ã‚¤ãƒ¼ãƒˆã‚«ã‚¦ãƒ³ãƒˆ
    fav_count = [] #ãŠæ°—ã«å…¥ã‚Šã‚«ã‚¦ãƒ³ãƒˆ
    
    if req.status_code == 200:
        search_timeline = json.loads(req.text)
        for tweet in search_timeline['statuses']:
            name.append(tweet['user']['name'])
            s_name.append(tweet['user']['screen_name'])
            tid.append(tweet['id'])
            text.append(tweet['text'])
            urls.append(tweet['entities']['urls'])
            geo.append(tweet['geo'])
            c_at.append(tweet['created_at'])
            ret_count.append(tweet['retweet_count'])
            fav_count.append(tweet['favorite_count'])
        new_max_id = tid[-1] #ãƒ©ã‚¹ãƒˆã«å–å¾—ã—ãŸãƒ„ã‚¤ãƒ¼ãƒˆIDã‚’å–å¾—
    else:
        print("ERROR: %d" % req.status_code)
    output = [name,s_name,tid,text,urls,geo,c_at,ret_count,fav_count,new_max_id]
    return(output)
```

ãƒ»ãƒªã‚¯ã‚¨ã‚¹ãƒˆã”ã¨ã«é–¢æ•°ã«æ¡ä»¶ã‚’æ¸¡ã—ã€ãƒ„ã‚¤ãƒ¼ãƒˆçµæœã‚’æ ¼ç´

ãƒ»å„ãƒªã‚¯ã‚¨ã‚¹ãƒˆã§å–å¾—ã—ãŸãƒ©ã‚¹ãƒˆã®ãƒ„ã‚¤ãƒ¼ãƒˆIDã‚’å–å¾—

ãƒ»æ¬¡ãƒªã‚¯ã‚¨ã‚¹ãƒˆã®max_idã«å‰ãƒªã‚¯ã‚¨ã‚¹ãƒˆã®ãƒ„ã‚¤ãƒ¼ãƒˆIDã‚’æ¸¡ã™ã€€ã®ç¹°ã‚Šè¿”ã—

```python
req_count = 0

l_name = []
l_s_name = []
l_tid = []
l_text = []
l_urls = []
l_geo = []
l_c_at = []
l_ret_count = []
l_fav_count = []
    
for x in range(180):
    params = {'q':q,'count':count,'geocode':geocode,
          'lang':lang,'result_type':result_type,
         'max_id':max_id,'until':until}
    output = get_tweet(url,params) #çµæœã®å–å¾—
    if output[7] == max_id:
        break
    else:
        l_name.extend(output[0])
        l_s_name.extend(output[1])
        l_tid.extend(output[2])
        l_text.extend(output[3])
        l_urls.extend(output[4])
        l_geo.extend(output[5])
        l_c_at.extend(output[6])
        l_ret_count.extend(output[7])
        l_fav_count.extend(output[8])
        max_id = output[9]-1
        req_count += 1
        print(str(x) + '::' + str(output[6][0]))
```

ãƒ»DataFrameä½œæˆ

```python
df = pd.DataFrame({
        'name':l_name, 
        'screen_name':l_s_name,
        'tweet_id':l_tid,
        'text':l_text,
        'url_info':l_urls,
        'geo_info':l_geo,
        'created_at':l_c_at,
        'retweet_count':l_ret_count,
        'favorite_count':l_fav_count})
df
```

ãƒ»ä½ç½®æƒ…å ±ã€urlã‚’æŒ¿å…¥

```python
df.to_csv("temp.csv")
df = pd.read_csv("temp.csv",encoding='utf_8',na_values='-')
```

```python
lat = []
lon = []
u = []

for row in df["geo_info"]:
    try:
        geo = ast.literal_eval(row)["coordinates"]
        lat.append(geo[0])
        lon.append(geo[1])
    except ValueError:
        lat.append("")
        lon.append("")

for row in df["url_info"]:
    try:
        url = ast.literal_eval(row)[0]
        u.append(url['url'])
    except IndexError:
        u.append("None")
    
df["lat"] = lat
df["lon"] = lon
df["url"] = u
```

ãƒ»created_atãŒç‰¹æ®Šãªå‹ã®ãŸã‚ã€datetimeå‹ãŠã‚ˆã³æ—¥æœ¬æ™‚é–“ã«å¤‰æ›

```python
created_at = []
for row in df["created_at"]:
    time_utc = time.strptime(row, '%a %b %d %H:%M:%S +0000 %Y')
    unix_time = calendar.timegm(time_utc)
    time_local = time.localtime(unix_time)
    japan_time = time.strftime("%Y-%m-%d %H:%M:%S", time_local)
    created_at.append(japan_time)
    
df["created_at_jp"] = created_at
df = df.sort_values('created_at_jp')
```

```python
df = df.drop("url_info", axis=1)
df = df.drop("url", axis=1)
df = df.drop("geo_info", axis=1)
df = df.drop("Unnamed: 0", axis=1)
df = df.drop("created_at", axis=1)
```
