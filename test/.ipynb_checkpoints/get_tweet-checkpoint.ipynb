1. キーワードでツイート取得

APIソース：https://developer.twitter.com/en/docs/tweets/search/api-reference/get-search-tweets.html


・OAuth認証

```python
import re
import pandas as pd
import ast
import json
import dask.dataframe as dd
from requests_oauthlib import OAuth1Session
import sys, time, calendar
```

```python
ck = "#############認証csafj;dlja####################"
cs = "###############認証コード##################"
at = "#################認証コード################"
ats = "###############認証コード#################"
twitter = OAuth1Session(ck, cs, at, ats)  #OAuth認証
```

・【検索条件】位置情報

'緯度, 軽度, 半径（キロ）'で指定

```python
geocode = '' #緯度経度なし
```

・【検索条件】キーワード、取得ツイート数、言語、結果のタイプ、日時条件、ツイートID条件

```python
q =  '' #検索したいキーワード
count = 100#条件：1～100 180req/15min
lang = 'ja'#言語
result_type = 'recent'#recent:最新、popular:人気のツイート
until = ''  #指定した日付まで取る
max_id = ''#取得したい最新のツイートID(ここから最新100件)

url = "https://api.twitter.com/1.1/search/tweets.json" #リクエストURL
```

・各リクエストごとにツイートを取得する関数を作成

```python
def get_tweet(url,params):
    req = twitter.get(url, params = params)
    name = [] #ユーザーのニックネーム 
    s_name = [] #ユーザー登録名
    tid = [] #ツイートID（ユニーク）
    text = [] #ツイート内容
    urls = [] #ツイートurl
    geo = [] #ツイートの位置情報
    c_at = [] #ツイート日時
    ret_count = [] #リツイートカウント
    fav_count = [] #お気に入りカウント
    
    if req.status_code == 200:
        search_timeline = json.loads(req.text)
        for tweet in search_timeline['statuses']:
            name.append(tweet['user']['name'])
            s_name.append(tweet['user']['screen_name'])
            tid.append(tweet['id'])
            text.append(tweet['text'])
            urls.append(tweet['entities']['urls'])
            geo.append(tweet['geo'])
            c_at.append(tweet['created_at'])
            ret_count.append(tweet['retweet_count'])
            fav_count.append(tweet['favorite_count'])
        new_max_id = tid[-1] #ラストに取得したツイートIDを取得
    else:
        print("ERROR: %d" % req.status_code)
    output = [name,s_name,tid,text,urls,geo,c_at,ret_count,fav_count,new_max_id]
    return(output)
```

・リクエストごとに関数に条件を渡し、ツイート結果を格納

・各リクエストで取得したラストのツイートIDを取得

・次リクエストのmax_idに前リクエストのツイートIDを渡す　の繰り返し

```python
req_count = 0

l_name = []
l_s_name = []
l_tid = []
l_text = []
l_urls = []
l_geo = []
l_c_at = []
l_ret_count = []
l_fav_count = []
    
for x in range(180):
    params = {'q':q,'count':count,'geocode':geocode,
          'lang':lang,'result_type':result_type,
         'max_id':max_id,'until':until}
    output = get_tweet(url,params) #結果の取得
    if output[7] == max_id:
        break
    else:
        l_name.extend(output[0])
        l_s_name.extend(output[1])
        l_tid.extend(output[2])
        l_text.extend(output[3])
        l_urls.extend(output[4])
        l_geo.extend(output[5])
        l_c_at.extend(output[6])
        l_ret_count.extend(output[7])
        l_fav_count.extend(output[8])
        max_id = output[9]-1
        req_count += 1
        print(str(x) + '::' + str(output[6][0]))
```

・DataFrame作成

```python
df = pd.DataFrame({
        'name':l_name, 
        'screen_name':l_s_name,
        'tweet_id':l_tid,
        'text':l_text,
        'url_info':l_urls,
        'geo_info':l_geo,
        'created_at':l_c_at,
        'retweet_count':l_ret_count,
        'favorite_count':l_fav_count})
df
```

・位置情報、urlを挿入

```python
df.to_csv("temp.csv")
df = pd.read_csv("temp.csv",encoding='utf_8',na_values='-')
```

```python
lat = []
lon = []
u = []

for row in df["geo_info"]:
    try:
        geo = ast.literal_eval(row)["coordinates"]
        lat.append(geo[0])
        lon.append(geo[1])
    except ValueError:
        lat.append("")
        lon.append("")

for row in df["url_info"]:
    try:
        url = ast.literal_eval(row)[0]
        u.append(url['url'])
    except IndexError:
        u.append("None")
    
df["lat"] = lat
df["lon"] = lon
df["url"] = u
```

```python
df
```

・created_atが特殊な型のため、datetime型および日本時間に変換

```python
created_at = []
for row in df["created_at"]:
    time_utc = time.strptime(row, '%a %b %d %H:%M:%S +0000 %Y')
    unix_time = calendar.timegm(time_utc)
    time_local = time.localtime(unix_time)
    japan_time = time.strftime("%Y-%m-%d %H:%M:%S", time_local)
    created_at.append(japan_time)
    
df["created_at_jp"] = created_at
df = df.sort_values('created_at_jp')
```

```python
df = df.drop("url_info", axis=1)
df = df.drop("url", axis=1)
df = df.drop("geo_info", axis=1)
df = df.drop("Unnamed: 0", axis=1)
df = df.drop("created_at", axis=1)
```
